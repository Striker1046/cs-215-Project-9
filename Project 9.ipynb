My project will be using data from a csv that has electric car registration data in Washington state

The questions that I will be answering for this project are:

What make of car is the most used in Washington? My guess - Tesla

I will also be creating 2 new columns to seperate out longitude and latitude for each car, which will be important for my visualization

```python
#importing the proper packages
import pandas as pd
import numpy as np
```

```python
#reading the csv file
cars_df = pd.read_csv("Electric_Vehicle_Population_Data.csv")

#There were some data that was null
cars_df = cars_df.dropna(subset = ["Vehicle Location"])
cars_df.reset_index()
```

```python
#Here is me trying to create new columns with the latitude and longitude
latitude_list = []
longitude_list = []

#Here is me testing out hjow to seperate the point into latitude and longitude

#point = "POINT (-120.56916 46.58514)"
#new_point = point.split("(", 1)[1]
#print(new_point.split(" ", 1)[1])

# this is the for loop that achieves the new columns, it takes awhile to run
point = str(cars_df["Vehicle Location"][0])
for i in range(len(cars_df)):
    point = str(cars_df.iloc[i]["Vehicle Location"]).split("(")[1]
    longitude = point.split(" ")[0]
    latitude = point.split(" ")[1].split(")")[0]
    latitude_list.append(latitude)
    longitude_list.append(longitude)
cars_df["Latitude"] = latitude_list
cars_df["Longitude"] = longitude_list
cars_df
```

```python
#A lot of the future calculations are somewhat similar, with only a few distinctions,
#so I decided to make a general function for all of them 

#This function returns a list with a string for the max count, and a string for the name of whatever had the max count
def max_name_and_count(df, column_name): 
    max_count = 0
    current_count = 0
    current_name = df.iloc[0][column_name]
    max_name = df.iloc[0][column_name]
    
    for value in range (len(df[column_name])):
        if (current_name == df.iloc[value][column_name]):
            current_count = current_count + 1
        else:
            current_name = df.iloc[value][column_name]
            current_count = 1
        if(current_count > max_count):
            max_name = df.iloc[value][column_name]
            max_count = current_count
    return [max_name, str(max_count)]
```

```python
#Here is where I answer the first question: 
#Where are the most electric vehicles registered in the state of Washington?

#I first needed to organize the dataframe by the city column
cars_df = cars_df.sort_values(by = ["City"], axis = 0)

#Using our function from earlier, max_values will be a list with the name of the max city and count of the max vehicle
max_values = max_name_and_count(cars_df, "City")

print("The city with the most electric vehicles registered in the state of Washington is " 
          + max_values[0] + " with a count of " + max_values[1])

#As expected, Seattle has the most electric cars registered at a count of 22006
print("That means that " + str(100*(int(max_values[1])/len(cars_df["City"]))) + "% of all electric vehicles " 
      + "in Washington state are registered in "+ max_values[0])
```

```python
#To confirm that the earlier number is accurate, we can use value_counts()
#I could have used value_counts from the start, but to show my skills as a computer scientist and coder, 
#I wanted to find the value in multiple ways
#I will be doing the same with Questions 2 and 3

print(cars_df["City"].value_counts())

#As you can see, seattle has the most electric cars with 22006, by far the most in Washington
```

```python
#Here is where I answer the second question: 
#What is the make of the car that is most used in the state of Washington
#We can use a similar algorithm to last time

#I first needed to organize the dataframe by the make column
cars_df = cars_df.sort_values(by = ["Make"], axis = 0)

#This is the function that we created earlier, whicxh calculates the max count and make
max_values = max_name_and_count(cars_df, "Make")

print("The city with the most electric vehicles registered in the state of Washington is " 
          + max_values[0] + " with a count of " + max_values[1])

#As expected, tesla is the car that is most popular in the state of Washington with 67118 cars registered
#There are more teslas in the state of washington than there are electric vehicles registered in Seattle

#I also wanted to find the percentage of total electric vehicles in Washington that are Teslas
print("That means that " + str(100*(int(max_values[1])/len(cars_df["Make"]))) + 
      "% of all electric vehicles in Washington state are "+ max_values[0])
```

```python
#To confirm that the earlier number is accurate, we can use value_counts()

print(cars_df["Make"].value_counts())

#As you can see, there are 57118 Teslas in the state of Washington, just like our previous count got.
```

```python
#We are going to start with getting data from Seattle and putting it into its own dataframe

seattle_df = cars_df[cars_df["City"] == "Seattle"]

#Now we are going to do a similar check to what we did before and see which model comes out on top

#First we need to sort
seattle_df = seattle_df.sort_values(by = ["Make"])
```

```python
#Here is the calculation for the most used make in Seattle
max_values = max_name_and_count(seattle_df, "Make")

print("The make of the most registered electric vehicle in Seattle is " 
          + max_values[0] + " with a count of " + max_values[1])

#These are the percentages
print("That means that " + str(100*(int(max_values[1])/len(seattle_df["Make"]))) + "% of all electric vehicles "
      + "in Seattle are "+ max_values[0])

#the about 42% here is slightly less than the 45% found in all of washington state, which makes sense because
#Seattle would probably have a wider selection of cars available than other, smaller cities and towns in Washington
```

```python
#Now to find the most popular model of that make, we need to do a little bit of more data wrangling

seattle_tesla_df = seattle_df[seattle_df["Make"] == "TESLA"]
seattle_tesla_df.head()
#This new dataframe has all the teslas in seattle, with only 9,241 rows 
```

```python
#This is where I find the most used model of tesla is used in Seattle

#First I need to sort by the Model Column
seattle_tesla_df = seattle_tesla_df.sort_values(by = ["Model"])
```

```python
#Now I use the function from earlier to retrun the max model and count

max_values = max_name_and_count(seattle_tesla_df, "Model")

#This print statement uses the list acquired from our previous function and displays the most used model of teslas
print("The model of the most registered electric model of Tesla in Seattle is " 
          + max_values[0] + " with a count of " + max_values[1])

#This is the percentage
print("That means that " + str(100*(int(max_values[1])/len(seattle_tesla_df["Model"]))) + "% of all Teslas "
      + "in Seattle are "+ max_values[0])
```
